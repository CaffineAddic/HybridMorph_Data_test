```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
import matplotlib.pyplot as plt
```

```python
path = '29092023/VRLab/CSV/'
# Get the files from the path provided in the OP
# files = glob.glob(os.path.join(path, "*.csv"))  # .rglob to get subdirectories
```

```python
real_loss = np.zeros((1000,1))
df = pd.read_csv(path+'Real_loss_val.csv')
df_numpy = df.to_numpy()
real_loss[0] = np.array(df.columns)
real_loss[1:] = df_numpy
```

```python
synth_loss = np.zeros((1000,1))
df = pd.read_csv(path+'Synth_loss_val.csv')
df_numpy = df.to_numpy()
synth_loss[0] = np.array(df.columns)
synth_loss[1:] = df_numpy
```

```python
plt.plot(real_loss)
plt.plot(synth_loss)
plt.legend(['real', 'synth'], loc='upper right')
plt.title('Validation Loss')
plt.ylabel('val loss')
plt.xlabel('epoch')
```
